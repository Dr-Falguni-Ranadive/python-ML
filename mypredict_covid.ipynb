```
import numpy as np
x = np.array([2.3,4.5,6.5,7.3,8.4,9.2,9.7])
y = np.array([2.7,3.8,4,6.9,7,8.2,9.1])
```

```
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```
#ordinary least square method
xmean=x.mean()
ymean=y.mean()
```

```
num=0
den=0
for i in range (len(x)):
  num += (x[i]-xmean)(y[i]-ymean)
  den += (x[i]-xmean)**2
m=num/den
c=ymean - m*xmean
```

```
import pandas as pd_world

path='covid-data-world 18May.csv'
df_world = pd_world.read_csv(path,parse_dates=['date'])
df_world.head()
```

```
from fbprophet import Prophet
dfindia.rename(columns={'date':'ds','total_cases':'y'},inplace=True)
dfindia.head()
```

```
imodel = Prophet(daily_seasonality=True, interval_width=0.95)
imodel.fit(dfindia)
```

```
future = imodel.make_future_dataframe(periods=7)
future.tail(107)
```

```
forecast = imodel.predict(future)
forecast.tail(7)
```
